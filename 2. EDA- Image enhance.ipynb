# Code to enhance the color feature of images [Brightness + Contrast+Sharpness]

```python
#Code to improve image Contract + brightness+Sharpness

import cv2
import os
import cv2
import os
from PIL import Image

from PIL import ImageEnhance

DataDir = "C:/Users/borka/OneDrive/Desktop/TF FULL/Enhance train and test to upload/Train/"
            


def Img(DataDir = "C:/Users/borka/OneDrive/Desktop/TF FULL/Enhance train and test to upload/Train"):
    return[os.path.join(DataDir, f) for f in os.listdir(DataDir) if f.endswith('jpg')]

filenames = Img()#code to read images path and write them back to directory


print('Total number of files to be processed are - ')
print(len(filenames))

for file in filenames:
    image = Image.open(file) # images are color imag
    enh_bri = ImageEnhance.Brightness(image)
    brightness = 1
    image_brightened = enh_bri.enhance(brightness) #Brightenup image
    image_brightened.save(file)
    
print('image Brightness enahancement completed..!!')
    
    
for file in filenames:
    image = Image.open(file) # images are color image
    enh_con = ImageEnhance.Contrast(image)
    contrast = 1
    image_contrasted = enh_con.enhance(contrast) #Contrast image 
    image_contrasted.save(file)

print('image Contract enahancement completed..!!')    
    
for file in filenames:
    image = Image.open(file) # images are color image
    
    enh_sha = ImageEnhance.Sharpness(image)
    sharpness = 3.0
    image_sharped = enh_sha.enhance(sharpness)  #Sharpen Image
    image_sharped.save(file)   
    
    
print('image Sharpening completed..!!')        


print('image Processing completed completed..!!')
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure as ex
import imageio
import sys

img = imageio.imread('C:/Users/borka/OneDrive/Desktop/TF FULL/EDA/299 299/Plastic/Plastic-35980.jpg')


def he(img):
    if(len(img.shape)==2):      # writing this one incase of need for conversion to gray scale image
        outImg = ex.equalize_hist(img[:,:])*255 
    elif(len(img.shape)==3):    #our actual image is in RGB
        outImg = np.zeros((img.shape[0],img.shape[1],3))
        for channel in range(img.shape[2]):
            outImg[:, :, channel] = ex.equalize_hist(img[:, :, channel])*255

    outImg[outImg>255] = 255
    outImg[outImg<0] = 0
    return outImg.astype(np.uint8)

result = he(img)
plt.imshow(result)
plt.show()

```
