```python
# Creating sample image dataset from each category
from glob import glob
import os
import numpy as np
import sys
from os import path
import random

from matplotlib.pyplot import figure, imshow, axis, title
from matplotlib.image import imread

Path = 'C:/Users/borka/OneDrive/Desktop/TF FULL/EDA/299 299/'
DirList = os.listdir(Path)

for i, dir in enumerate(DirList): 
    if dir != "Test" and dir != "Train": 
        fig = figure(figsize=(9, 9))
        imageList = glob(Path + "/" + dir + '/*.jpg')
        # print (imageList)
        index = 0
        random.shuffle(imageList)
        for file in imageList:
            if index < 5:
                a = fig.add_subplot(1, 5, index+1)
                image = imread(file)
                imshow(image)
                if index == 0:
                    title(dir)
                axis('off')
                index = index + 1
```
